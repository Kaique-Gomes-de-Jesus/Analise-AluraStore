### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)


loja.head()
```

#1. Análise do faturamento


```python
#Faturamento de cada loja da AluraStore
loja = sum(loja["Preço"])
loja2 = sum(loja2["Preço"])
loja3 = sum(loja3["Preço"])
loja4 = sum(loja4["Preço"])
print(f"""
    Faturamento da loja 1: {loja:.2f}
    Faturamento da loja 2: {loja2:.2f}
    Faturamento da loja 3: {loja3:.2f}
    Faturamento da loja 4: {loja4:.2f}
      """)
```

```python


fig, ax = plt.subplots() #Criando a figura do quadro e os eixos

lojas = ['loja1', 'loja2', 'loja3', 'loja4'] #Nome das lojas
faturamento = [loja, loja2, loja3, loja4] #Faturamento das lojas
cores = ['green', 'orange', 'blue', 'red'] #Cores das barras

ax.bar(lojas, faturamento, label=lojas, color=cores) #Criando o gráfico com os dados nome das lojas, faturamento, legenda e cores das barras

ax.set_ylabel('Faturamento') #Texto eixo y
ax.set_xlabel('Lojas') #Texto eixo x
ax.set_title('Faturamento das lojas') #Título do gráfico
ax.legend(title='Cores das lojas') #Título da legenda

plt.show() #Exibindo o gráfico na tela
```

# 2. Vendas por Categoria


```python
#Dicionário para armazenar lojas e categorias
categorias = {
    "loja1":loja['Categoria do Produto'].value_counts(), #Somando quantas vezes aparece cada produto
    "loja2":loja2['Categoria do Produto'].value_counts(),
    "loja3":loja3['Categoria do Produto'].value_counts(),
    "loja4":loja4['Categoria do Produto'].value_counts()
}
print(categorias)
```

```python
for loja, dados in categorias.items(): #Para cada loja e dados/qtd vendidas no dicionário categorias
    ax = dados.plot(kind='bar', color='orange') #Criando gráfico de barras com os dados

    plt.title(f"Categorias Mais Vendidas - {loja}")
    plt.xlabel('Categoria')
    plt.ylabel('Quantidade Vendida')
    plt.box(False) #Tirando o quadrado em volta do gráfico
    # Adicionando os valores em cima de cada barra
    for i, v in enumerate(dados): #Para barra e valor nos dados
        plt.text(i, v + 0.5, str(v), ha='center') #Escreve na barra, no valor, o valor

    plt.show() #Mostra o gráfico
```

# 3. Média de Avaliação das Lojas

```python
medias = [] #Lista para armazenar media de cada loja
lojas = [loja, loja2, loja3, loja4]
for i,loja in enumerate(lojas): #Pegando índice e dados das lojas
    media = sum(loja['Avaliação da compra']) / len(loja['Avaliação da compra']) #Cálculo média
    medias.append(media)
    print(f"A média das lojas {i + 1}: {media:.3f} ")
```

```python
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
medias = medias
barras = plt.barh(lojas, medias) #Criando gráfico de barras na horizontal

plt.bar_label(barras, padding=3, fontsize = 10, fmt='%.2f') #Colocando valores nas barras


plt.xticks([]) #Deixando o eixo X vazio
plt.box(False)
plt.title("Média de avaliação por loja")

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
#Dicionário para armazenar lojas e produtos
categorias = {
    "Loja 1":loja['Produto'],
    "Loja 2":loja2['Produto'],
    "Loja 3":loja3['Produto'],
    "Loja 4":loja4['Produto']
}
#Laço para percorrer nome da loja e produtos
for nome, produtos in categorias.items():
    print(f"Produtos mais vendidos na {nome}: {produtos.value_counts().head(3)}\n") #Somando qtd de cada produto e exibindo as 3 primeiras linhas
    print(f"Produtos menos vendidos na {nome}: {produtos.value_counts().tail(3)}\n") #Somando qtd de cada produto e exibindo as 3 últimas linhas



```

```python
#Dados e nomes das lojas
lojas = [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

#Listas para armanezar valores dos produtos mais e menos vendidos
mais_vendidos = []
menos_vendidos = []
#Listas para armanezar nomes dos produtos mais e menos vendidos
mais_vendidos_nome = []
menos_vendidos_nome = []

#Laço para percorrer os dados de cada loja
for loja in lojas:
    contagem = loja['Produto'].value_counts()
    mais_vendidos.append(contagem.max()) #Adicionando a quantidade do produto mais vendido na lista
    menos_vendidos.append(contagem.min()) #Adicionando a quantidade do produto menos vendido na lista
    mais_vendidos_nome.append(contagem.idxmax()) #Adicionando o nome do produto mais vendido na lista
    menos_vendidos_nome.append(contagem.idxmin()) #Adicionando o nome do produto menos vendido na lista

x = np.arange(len(nomes_lojas))  # Posições no eixo X
largura = 0.35  # Largura das barras

#Criando rótulos com o nome e a qtd dos produtos vendidos com List comprehension
labels_mais = [f"{nome} ({qtd})" for nome, qtd in zip(mais_vendidos_nome, mais_vendidos)]
labels_menos = [f"{nome} ({qtd})" for nome, qtd in zip(menos_vendidos_nome, menos_vendidos)]

#Criando figura e eixos
fig, ax = plt.subplots(figsize=(10,5))
#Criando gráfico de barras
barras1 = ax.bar(x - largura/2, mais_vendidos, largura, label='Mais Vendido', color = "green")
barras2 = ax.bar(x + largura/2, menos_vendidos, largura, label='Menos Vendido', color = "red")

# Adicionando rótulos nas barras
ax.bar_label(barras1, labels=labels_mais, padding=3)
ax.bar_label(barras2, labels=labels_menos, padding=3)

#Configurações do gráfico
ax.set_ylabel('Quantidade Vendida')
ax.set_title('Produtos Mais e Menos Vendidos por Loja')
ax.set_xticks(x)
ax.set_xticklabels(nomes_lojas)
ax.legend()

plt.show()
```

# 5. Frete Médio por Loja


```python
#Lista para armazenar os fretes de cada loja
medias_frete = []
lojas = [loja, loja2, loja3, loja4]
for i,loja in enumerate(lojas):
    media = sum(loja['Frete']) / len(loja['Frete']) #Média do frete de cada loja
    medias_frete.append(media)
    print(f"A média de frete da loja {i + 1} é de: R${media:.2f} ")
```

```python
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
medias_frete = medias_frete

#Criando figura
plt.figure(figsize=(6,6))
#Criando gráfico de pizza
plt.pie(medias_frete, labels=lojas, autopct="%1.1f%%", startangle=90)

plt.title("Média de frete por loja")
#Deixando o gráfico circular
plt.axis("equal")
plt.show()
```
